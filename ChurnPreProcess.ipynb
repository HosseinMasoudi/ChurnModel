```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
DataFrame = pd.read_csv("Churn_Modelling.csv")
```

```python
DataFrame = DataFrame.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder

LE = LabelEncoder()
DataFrame['Gender'] = LE.fit_transform(DataFrame['Gender'])  # Female=0, Male=1
DataFrame = pd.get_dummies(DataFrame, columns=['Geography'], dtype=int)
```

```python
DataFrame.sample(20)
```

```python
DataFramePreprocessed = DataFrame.copy()
DataFramePreprocessed.to_csv('churn_preprocessed.csv', index=False)
```
